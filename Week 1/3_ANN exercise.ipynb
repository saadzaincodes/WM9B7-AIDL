### 2. Extend the model to use two layers instead of one

This part is about adding more layers to get more powerful models. We can't simply stack linear operations on top of each other as they would simplify into one linear operation, so we need to sandwitch non-linear activation functions inbetween. 

The layers that are not input or output layers are known as hidden layers, as we do not typically see the values stored in these neurons. Generally speaking, when going from one layer to the next we want the number of neurons to stay the same or decrease. More neurons means the layer can store more information, and we don't need our model to store more information than we were given in the first place.
