## 2. Tensors — The Fundamental Data Structure

A **tensor** is PyTorch's equivalent of a NumPy array. Everything in deep learning — inputs, weights, gradients, outputs — is stored as a tensor.

- A **scalar** is a 0-dimensional tensor
- A **vector** is a 1-dimensional tensor
- A **matrix** is a 2-dimensional tensor
- An **image batch** is typically a 4-dimensional tensor: `(batch, channels, height, width)`

### 2.1 Creating Tensors

````python
# From Python lists
a = torch.tensor([1, 2, 3])
print(f'From list:   {a},  shape: {a.shape},  dtype: {a.dtype}')

# From a 2D list (matrix)
M = torch.tensor([[1, 2, 3],
                   [4, 5, 6]])
print(f'Matrix shape: {M.shape}  (2 rows, 3 columns)')

# Specifying the data type — float32 is the standard for deep learning
x = torch.tensor([1.0, 2.0, 3.0])
print(f'Float tensor: {x},  dtype: {x.dtype}')

# You can also specify explicitly
x_float = torch.tensor([1, 2, 3], dtype=torch.float32)
print(f'Explicit float: {x_float},  dtype: {x_float.dtype}')
````

### 2.2 Common Tensor Creation Functions

These are used constantly — for initialising weights, creating masks, etc.

````python
# Zeros and ones
print(f'zeros(2,3):\n{torch.zeros(2, 3)}')
print(f'\nones(3):\n{torch.ones(3)}')

# Random tensors (most common for weight initialisation)
print(f'\nrandn(2,3) — random from N(0,1):\n{torch.randn(2, 3)}')
print(f'\nrand(2,3) — random from U(0,1):\n{torch.rand(2, 3)}')

# Range
print(f'\narange(0, 10, 2): {torch.arange(0, 10, 2)}')

# linspace
print(f'\nlinspace(0, 1, 5): {torch.linspace(0, 1, 5)}')

# Identity matrix
print(f'\neye(3):\n{torch.eye(3)}')

# Full (fill with a value)
print(f'\nfull((2,3), fill_value=7):\n{torch.full((2, 3), 7.0)}')
````

### 3.2 Matrix Multiplication

This is the core operation in neural networks: `z = W @ x + b`

````python
# Matrix multiplication: use @ or torch.mm or torch.matmul
W = torch.randn(2, 3)  # 2 neurons, 3 inputs
x = torch.randn(3, 1)  # 3 features, 1 sample

z = W @ x  # (2,3) @ (3,1) = (2,1)
print(f'W shape: {W.shape}, x shape: {x.shape}')
print(f'W @ x shape: {z.shape}')
print(f'W @ x =\n{z}')

# Batch matrix multiplication: matmul handles batches
X = torch.randn(3, 4)  # 3 features, 4 samples (a batch)
Z = W @ X               # (2,3) @ (3,4) = (2,4) — all 4 samples at once!
print(f'\nBatch: W @ X shape: {Z.shape}  (2 outputs for 4 samples)')
````

## 4. NumPy ↔ PyTorch Conversion

You will often need to convert between NumPy (for data loading, plotting) and PyTorch (for training).

````python
# NumPy → PyTorch
np_array = np.array([1.0, 2.0, 3.0])
torch_tensor = torch.from_numpy(np_array)
print(f'NumPy:   {np_array}, type: {type(np_array)}')
print(f'Tensor:  {torch_tensor}, type: {type(torch_tensor)}')

# PyTorch → NumPy
back_to_numpy = torch_tensor.numpy()
print(f'Back:    {back_to_numpy}, type: {type(back_to_numpy)}')

# WARNING: they share memory! Changing one changes the other.
np_array[0] = 999
print(f'\nAfter modifying NumPy array:')
print(f'NumPy:  {np_array}')
print(f'Tensor: {torch_tensor}  ← also changed!')

# To avoid shared memory, use .clone()
safe_tensor = torch.from_numpy(np.array([1.0, 2.0])).clone()
print(f'\nCloned tensor is independent: {safe_tensor}')
````

**When to convert:**
- **NumPy → Tensor**: before feeding data into a model
- **Tensor → NumPy**: for plotting with matplotlib (`tensor.detach().numpy()`)
- Use `.detach()` first if the tensor has gradients (more on this next)

## 5. Autograd — Automatic Differentiation

This is PyTorch's **killer feature**. When you set `requires_grad=True` on a tensor, PyTorch tracks every operation on it. When you call `.backward()`, it automatically computes all the gradients using the chain rule.

This is what makes training neural networks possible — you define the forward pass, and PyTorch figures out the backward pass for you.

### 5.1 Basic Example

If $y = x^2$ and $x = 3$, then $\frac{dy}{dx} = 2x = 6$.

````python
# Create a tensor that tracks gradients
x = torch.tensor(3.0, requires_grad=True)

# Forward pass: compute y = x^2
y = x ** 2
print(f'x = {x.item():.1f}')
print(f'y = x² = {y.item():.1f}')

# Backward pass: compute dy/dx automatically
y.backward()

# The gradient is stored in x.grad
print(f'dy/dx = 2x = 2({x.item():.1f}) = {x.grad.item():.1f}')
````

### What Just Happened?

1. We created `x = 3.0` with `requires_grad=True` — telling PyTorch to track operations.
2. We computed `y = x²` — PyTorch recorded this operation in a computation graph.
3. We called `y.backward()` — PyTorch traversed the graph backwards, computing $\frac{dy}{dx} = 2x = 6$.
4. The result was stored in `x.grad`.

This is exactly what happens during neural network training, except with millions of parameters instead of one.

### 5.2 A More Realistic Example: Gradient Descent Step

````python
# Simulating one step of gradient descent
# We want to find x that minimises f(x) = (x - 5)²

x = torch.tensor(0.0, requires_grad=True)  # start at x=0
lr = 0.1  # learning rate

print(f'{"Step":>4s}  {"x":>8s}  {"f(x)":>8s}  {"grad":>8s}')
print('-' * 35)

for step in range(10):
    # Forward pass
    loss = (x - 5) ** 2

    # Backward pass
    loss.backward()

    # Gradient descent update (must be inside no_grad!)
    with torch.no_grad():
        x -= lr * x.grad
        x.grad.zero_()  # IMPORTANT: reset gradients

    print(f'{step:4d}  {x.item():8.4f}  {loss.item():8.4f}  {2*(x.item()-5):8.4f}')

print(f'\nFinal x = {x.item():.4f} (should approach 5.0)')
````

### Key Points About Autograd

- **`requires_grad=True`**: tells PyTorch to track operations on this tensor.
- **`.backward()`**: computes gradients via backpropagation.
- **`.grad`**: stores the computed gradient.
- **`.grad.zero_()`**: resets gradients to zero. **You must do this before each backward pass** or gradients accumulate!
- **`torch.no_grad()`**: temporarily disables gradient tracking. Used during weight updates and inference.
- **`.detach()`**: creates a tensor that doesn't track gradients. Used when converting to NumPy for plotting.

## 6. Building Models with `nn.Module`

In practice, you don't define weights manually — you use `torch.nn.Module`, which manages weights, biases, and gradient tracking for you.

### 6.1 `nn.Linear` — The Fully Connected Layer

`nn.Linear(in_features, out_features)` computes $y = Wx + b$.

It automatically creates the weight matrix $W$ and bias vector $b$ with the correct shapes.

````python
import torch.nn as nn

# A single linear layer: 3 inputs → 2 outputs
layer = nn.Linear(in_features=3, out_features=2)

print(f'Weight shape: {layer.weight.shape}  (out_features × in_features)')
print(f'Bias shape:   {layer.bias.shape}    (out_features)')
print(f'Weight:\n{layer.weight.data.round(decimals=3)}')
print(f'Bias: {layer.bias.data.round(decimals=3)}')

# Pass data through the layer
x = torch.tensor([[1.0, 2.0, 3.0]])  # one sample, 3 features — shape (1, 3)
output = layer(x)
print(f'\nInput shape:  {x.shape}')
print(f'Output shape: {output.shape}  (1 sample, 2 outputs)')
print(f'Output: {output.data.round(decimals=3)}')

# Batch of 4 samples
X = torch.randn(4, 3)  # 4 samples, 3 features
output = layer(X)
print(f'\nBatch input:  {X.shape}')
print(f'Batch output: {output.shape}  (4 samples, 2 outputs)')
````

**Important note on shapes:** In PyTorch's `nn.Linear`, data has shape `(batch, features)` — each **row** is one sample. This is different from the column convention we used in the NumPy notebook, but it is the PyTorch standard.

### 6.2 Defining a Neural Network

We define networks by subclassing `nn.Module`. This is the pattern you will use for **every** model in this module — ANNs, CNNs, RNNs, GANs, everything.

````python
class SimpleNetwork(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super().__init__()
        # Define layers
        self.hidden = nn.Linear(input_size, hidden_size)
        self.output = nn.Linear(hidden_size, output_size)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        # Define how data flows through the network
        x = self.hidden(x)    # linear transformation
        x = self.sigmoid(x)   # activation function
        x = self.output(x)    # linear transformation
        x = self.sigmoid(x)   # output activation
        return x

# Create the model
model = SimpleNetwork(input_size=2, hidden_size=4, output_size=1)

print('Model architecture:')
print(model)

# Count parameters
total_params = sum(p.numel() for p in model.parameters())
print(f'\nTotal parameters: {total_params}')

# List all parameter shapes
for name, param in model.named_parameters():
    print(f'  {name}: {param.shape}')
````

### What Just Happened?

1. We defined a class that inherits from `nn.Module`.
2. In `__init__`, we defined the layers (this is where weights are created).
3. In `forward`, we defined how data flows through the layers.
4. PyTorch automatically handles the backward pass — we never define it!

**This two-method pattern (`__init__` + `forward`) is used for ALL neural networks in PyTorch.**

### 6.3 Using `nn.Sequential` — A Shortcut

````python
# nn.Sequential is a convenience for simple feed-forward networks
# It chains layers together automatically

model_seq = nn.Sequential(
    nn.Linear(2, 4),
    nn.Sigmoid(),
    nn.Linear(4, 1),
    nn.Sigmoid()
)

print('Sequential model:')
print(model_seq)

# Forward pass — just call the model
x = torch.randn(3, 2)  # 3 samples, 2 features
output = model_seq(x)
print(f'\nInput:  {x.shape}')
print(f'Output: {output.shape}')
print(f'Predictions:\n{output.data.round(decimals=4)}')
````

**When to use which:**
- **`nn.Sequential`**: for simple, straight-through networks (no branching).
- **Custom `nn.Module`**: when you need skip connections, multiple inputs/outputs, conditional logic, or anything more complex.

## 7. Loss Functions and Optimisers

### 7.1 Loss Functions

PyTorch provides all common loss functions in `torch.nn`.

````python
# Binary Cross-Entropy — for binary classification (sigmoid output)
bce = nn.BCELoss()
predictions = torch.tensor([0.9, 0.1, 0.8])  # model outputs (after sigmoid)
targets = torch.tensor([1.0, 0.0, 1.0])       # true labels
loss = bce(predictions, targets)
print(f'BCELoss: {loss.item():.4f}')

# BCE with Logits — combines sigmoid + BCE (more numerically stable)
bce_logits = nn.BCEWithLogitsLoss()
logits = torch.tensor([2.0, -2.0, 1.5])  # raw model outputs (before sigmoid)
loss = bce_logits(logits, targets)
print(f'BCEWithLogitsLoss: {loss.item():.4f}')

# Cross-Entropy — for multi-class classification (softmax output)
ce = nn.CrossEntropyLoss()
logits = torch.tensor([[2.0, 0.5, 0.1],    # sample 1: class 0 has highest score
                        [0.1, 2.5, 0.3]])   # sample 2: class 1 has highest score
targets = torch.tensor([0, 1])               # true classes
loss = ce(logits, targets)
print(f'CrossEntropyLoss: {loss.item():.4f}')

# MSE — for regression
mse = nn.MSELoss()
predictions = torch.tensor([2.5, 0.5, 1.0])
targets = torch.tensor([3.0, 0.0, 1.0])
loss = mse(predictions, targets)
print(f'MSELoss: {loss.item():.4f}')
````

**Which loss to use:**

| Task | Output Activation | Loss Function |
|:---|:---|:---|
| Binary classification | Sigmoid | `BCELoss` or `BCEWithLogitsLoss` |
| Multi-class classification | None (raw logits) | `CrossEntropyLoss` (includes softmax) |
| Regression | None (linear) | `MSELoss` |

### 7.2 Optimisers

Optimisers update the model's weights based on the computed gradients.

````python
import torch.optim as optim

model = nn.Linear(3, 1)  # simple model for demonstration

# SGD (Stochastic Gradient Descent)
optimizer_sgd = optim.SGD(model.parameters(), lr=0.01)

# Adam (adaptive learning rate — the most popular choice)
optimizer_adam = optim.Adam(model.parameters(), lr=0.001)

print(f'SGD optimizer:  lr={optimizer_sgd.defaults["lr"]}')
print(f'Adam optimizer: lr={optimizer_adam.defaults["lr"]}')

# The key methods:
# optimizer.zero_grad()  — reset gradients to zero
# optimizer.step()       — update weights using gradients
````

## 8. The Training Loop — The Most Important Pattern

**Every** deep learning training script follows this pattern:

```python
for epoch in range(num_epochs):
    # 1. Forward pass
    predictions = model(inputs)

    # 2. Compute loss
    loss = loss_fn(predictions, targets)

    # 3. Backward pass
    optimizer.zero_grad()   # reset gradients
    loss.backward()         # compute gradients

    # 4. Update weights
    optimizer.step()        # apply gradients
```

Let's see this in action with a complete example.

### 8.1 Complete Example: Learning XOR with a Neural Network

````python
# === DATA ===
X = torch.tensor([[0, 0], [0, 1], [1, 0], [1, 1]], dtype=torch.float32)
y = torch.tensor([[0], [1], [1], [0]], dtype=torch.float32)

# === MODEL ===
torch.manual_seed(42)
model = nn.Sequential(
    nn.Linear(2, 4),
    nn.Sigmoid(),
    nn.Linear(4, 1),
    nn.Sigmoid()
)

# === LOSS AND OPTIMISER ===
criterion = nn.BCELoss()
optimizer = optim.SGD(model.parameters(), lr=2.0)

# === TRAINING LOOP ===
losses = []

for epoch in range(5000):
    # 1. Forward pass
    y_pred = model(X)

    # 2. Compute loss
    loss = criterion(y_pred, y)
    losses.append(loss.item())

    # 3. Backward pass
    optimizer.zero_grad()
    loss.backward()

    # 4. Update weights
    optimizer.step()

    if epoch % 1000 == 0 or epoch == 4999:
        acc = ((y_pred >= 0.5).float() == y).float().mean()
        print(f'Epoch {epoch:5d}: loss = {loss.item():.4f}, accuracy = {acc.item():.0%}')
````

Let's understand what happened in each epoch:

1. **Forward pass** (`model(X)`): data flows through the network, producing predictions.
2. **Loss computation** (`criterion(y_pred, y)`): measures how wrong the predictions are.
3. **`optimizer.zero_grad()`**: clears old gradients. Without this, gradients would accumulate!
4. **`loss.backward()`**: computes gradients of the loss with respect to every parameter.
5. **`optimizer.step()`**: updates every parameter using its gradient.

````python
# Verify the trained model
with torch.no_grad():  # no need to track gradients for inference
    predictions = model(X)

print('Final predictions:')
for i in range(4):
    pred_class = 1 if predictions[i].item() >= 0.5 else 0
    print(f'  Input: {X[i].tolist()} → P(y=1) = {predictions[i].item():.4f} → class {pred_class} (true: {int(y[i].item())})')
````

````python
# Plot the training loss
import matplotlib.pyplot as plt

plt.rcParams.update({
    'figure.figsize': (9, 4), 'axes.facecolor': '#0F172A',
    'figure.facecolor': '#0F172A', 'text.color': '#E2E8F0',
    'axes.labelcolor': '#E2E8F0', 'xtick.color': '#94A3B8',
    'ytick.color': '#94A3B8', 'axes.edgecolor': '#334155',
    'grid.color': '#334155', 'grid.alpha': 0.5,
})

plt.plot(losses, color='#F59E0B', linewidth=1.5)
plt.xlabel('Epoch'); plt.ylabel('BCE Loss')
plt.title('XOR Training Loss', fontweight='bold')
plt.grid(True); plt.tight_layout(); plt.show()
````

## 9. `Dataset` and `DataLoader`

For real projects, you use `Dataset` to define your data and `DataLoader` to load it in mini-batches. This is essential for training on large datasets that don't fit in memory.

### 9.1 `TensorDataset` — The Simple Case

````python
from torch.utils.data import TensorDataset, DataLoader

# Create a dataset
X = torch.randn(100, 3)  # 100 samples, 3 features
y = torch.randint(0, 2, (100, 1)).float()  # 100 binary labels

dataset = TensorDataset(X, y)
print(f'Dataset size: {len(dataset)}')
print(f'First sample: X={dataset[0][0].round(decimals=3)}, y={dataset[0][1]}')

# Create a DataLoader for mini-batch training
loader = DataLoader(dataset, batch_size=16, shuffle=True)

# Iterate through one epoch
for batch_idx, (X_batch, y_batch) in enumerate(loader):
    if batch_idx < 3:  # show first 3 batches
        print(f'Batch {batch_idx}: X shape = {X_batch.shape}, y shape = {y_batch.shape}')

print(f'\nTotal batches per epoch: {len(loader)}  (100 samples / 16 per batch)')
````

### 9.2 Custom `Dataset` — For Complex Data

````python
from torch.utils.data import Dataset

class CustomDataset(Dataset):
    def __init__(self, num_samples=200):
        # Generate synthetic data
        self.X = torch.randn(num_samples, 5)
        self.y = (self.X.sum(dim=1) > 0).float().unsqueeze(1)

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]

# Usage
dataset = CustomDataset(200)
loader = DataLoader(dataset, batch_size=32, shuffle=True)

print(f'Dataset size: {len(dataset)}')
print(f'Batches per epoch: {len(loader)}')

# Get one batch
X_batch, y_batch = next(iter(loader))
print(f'Batch shapes: X={X_batch.shape}, y={y_batch.shape}')
````

## 10. Key Layer Types for the Module

This section previews the layers you will use throughout the module. You don't need to understand the theory yet — just learn the PyTorch commands and what shapes they expect.

### 10.1 Fully Connected / Linear Layer (for ANN, MLP)

$$y = Wx + b$$

Used in: Perceptrons, ANNs, MLPs, the classification heads of CNNs and RNNs.

````python
# nn.Linear(in_features, out_features)
fc = nn.Linear(10, 5)  # 10 inputs → 5 outputs

x = torch.randn(3, 10)  # batch of 3, each with 10 features
out = fc(x)

print(f'Input:  {x.shape}')
print(f'Output: {out.shape}  (3 samples, 5 outputs)')
print(f'Params: W={fc.weight.shape}, b={fc.bias.shape}')
````

### 10.2 Convolutional Layer (for CNNs)

`nn.Conv2d` slides a small filter across a 2D image, detecting patterns like edges, textures, and shapes.

Input shape: `(batch, channels, height, width)`

````python
# nn.Conv2d(in_channels, out_channels, kernel_size)
conv = nn.Conv2d(in_channels=3, out_channels=16, kernel_size=3, padding=1)

# Simulate a batch of 2 RGB images, each 32x32 pixels
images = torch.randn(2, 3, 32, 32)  # (batch, channels, H, W)
out = conv(images)

print(f'Input:  {images.shape}  (2 images, 3 channels RGB, 32×32)')
print(f'Output: {out.shape}  (2 images, 16 feature maps, 32×32)')
print(f'Kernel: {conv.weight.shape}  (16 filters, 3 channels, 3×3)')
````

**Common companion layers for CNNs:**

````python
# MaxPool2d — reduces spatial dimensions
pool = nn.MaxPool2d(kernel_size=2, stride=2)
x = torch.randn(1, 16, 32, 32)
print(f'Before pooling: {x.shape}')
print(f'After pooling:  {pool(x).shape}  (32→16: halved!)')

# BatchNorm2d — normalises activations (stabilises training)
bn = nn.BatchNorm2d(16)
print(f'\nBatchNorm input:  {x.shape}')
print(f'BatchNorm output: {bn(x).shape}  (shape unchanged, values normalised)')

# Dropout — randomly zeros elements during training (prevents overfitting)
drop = nn.Dropout(p=0.5)
x_small = torch.ones(1, 10)
print(f'\nBefore dropout: {x_small}')
print(f'After dropout:  {drop(x_small)}  (some elements zeroed, rest scaled up)')
````

### 10.3 Recurrent Layers (for RNNs and LSTMs)

These process **sequential data** (text, time series, audio) one time step at a time, maintaining a hidden state that captures context.

Input shape: `(batch, sequence_length, features)`

### RNN

````python
# nn.RNN(input_size, hidden_size, num_layers, batch_first=True)
rnn = nn.RNN(input_size=10, hidden_size=20, num_layers=1, batch_first=True)

# Simulate a batch of 3 sequences, each with 5 time steps, 10 features per step
x = torch.randn(3, 5, 10)  # (batch, seq_len, features)

output, h_n = rnn(x)

print(f'Input:       {x.shape}       (3 sequences, 5 time steps, 10 features)')
print(f'Output:      {output.shape}  (3 sequences, 5 time steps, 20 hidden units)')
print(f'Final state: {h_n.shape}     (1 layer, 3 sequences, 20 hidden units)')
print(f'\nThe output contains the hidden state at every time step.')
print(f'h_n contains only the final hidden state (useful for classification).')
````

### LSTM (Long Short-Term Memory)

````python
# LSTM is like RNN but with a cell state that helps remember long-range dependencies
lstm = nn.LSTM(input_size=10, hidden_size=20, num_layers=2, batch_first=True)

x = torch.randn(3, 5, 10)  # same input as before

output, (h_n, c_n) = lstm(x)

print(f'Input:       {x.shape}')
print(f'Output:      {output.shape}   (hidden state at each time step)')
print(f'h_n:         {h_n.shape}     (final hidden state, 2 layers)')
print(f'c_n:         {c_n.shape}     (final cell state, 2 layers — unique to LSTM)')
````

### 10.4 Embedding Layer (for NLP)

````python
# nn.Embedding maps integer indices to dense vectors
# vocab_size=1000 words, each mapped to a 64-dimensional vector
embed = nn.Embedding(num_embeddings=1000, embedding_dim=64)

# Simulate a batch of 2 sentences, each 10 words (represented as integer IDs)
word_ids = torch.randint(0, 1000, (2, 10))
embedded = embed(word_ids)

print(f'Word IDs:  {word_ids.shape}   (2 sentences, 10 words each)')
print(f'Embedded:  {embedded.shape}  (2 sentences, 10 words, 64-dim vectors)')
print(f'\nEach integer word ID is now a 64-dimensional vector.')
print(f'These vectors are learned during training.')
````

### 10.5 Building a Complete CNN (Preview)

````python
class SimpleCNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.features = nn.Sequential(
            nn.Conv2d(1, 16, kernel_size=3, padding=1),  # 1 input channel (grayscale)
            nn.ReLU(),
            nn.MaxPool2d(2),  # 28→14
            nn.Conv2d(16, 32, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),  # 14→7
        )
        self.classifier = nn.Sequential(
            nn.Flatten(),         # 32*7*7 = 1568
            nn.Linear(32*7*7, 128),
            nn.ReLU(),
            nn.Linear(128, 10),   # 10 classes (digits 0-9)
        )

    def forward(self, x):
        x = self.features(x)
        x = self.classifier(x)
        return x

cnn = SimpleCNN()
print(cnn)

# Test with a batch of MNIST-like images
images = torch.randn(4, 1, 28, 28)  # 4 grayscale 28×28 images
output = cnn(images)
print(f'\nInput:  {images.shape}')
print(f'Output: {output.shape}  (4 samples, 10 class scores)')
print(f'Total params: {sum(p.numel() for p in cnn.parameters()):,}')
````

## 11. Saving and Loading Models

After training, you save the model so you can use it later without retraining.

````python
# Save model weights
model = nn.Sequential(nn.Linear(2, 4), nn.Sigmoid(), nn.Linear(4, 1))
torch.save(model.state_dict(), '/tmp/model_weights.pth')
print('Model saved!')

# Load model weights
model_loaded = nn.Sequential(nn.Linear(2, 4), nn.Sigmoid(), nn.Linear(4, 1))
model_loaded.load_state_dict(torch.load('/tmp/model_weights.pth', weights_only=True))
model_loaded.eval()  # set to evaluation mode (disables dropout etc.)
print('Model loaded!')

# Verify they produce the same output
x = torch.randn(1, 2)
with torch.no_grad():
    out1 = model(x)
    out2 = model_loaded(x)
print(f'\nOriginal output: {out1.item():.6f}')
print(f'Loaded output:   {out2.item():.6f}')
print(f'Match: {torch.allclose(out1, out2)}')
````

**Key methods:**
- `model.train()` — sets model to training mode (dropout and batchnorm active)
- `model.eval()` — sets model to evaluation mode (dropout disabled, batchnorm uses running stats)
- `torch.no_grad()` — disables gradient computation (use during inference for speed)

### The Universal Training Pattern
```python
model = MyModel().to(device)
criterion = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

for epoch in range(num_epochs):
    for X_batch, y_batch in train_loader:
        X_batch, y_batch = X_batch.to(device), y_batch.to(device)
        y_pred = model(X_batch)
        loss = criterion(y_pred, y_batch)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
```
