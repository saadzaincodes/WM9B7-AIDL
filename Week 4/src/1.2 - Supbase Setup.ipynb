## Step 4: Create a New Project

Now create a new project. Fill in your project name, set a **strong database password** (you'll need this later!), and select your preferred region.

Add your database password to the `Week 4/.env` file immediately so you don't forget it:


![Project Dashboard](../assets/supbase_setup_images/Supbase_5.png)

## Step 5.1: Click "Connect" to Get Database Details

Click the "Connect" button in the top right to access your database connection information.

![Click Connect](../assets/supbase_setup_images/Supbase_5_1.png)

## Step 6: Choose Direct Connection String

Select "Direct" > "Connection string" to get your database connection details.

![Connection Dialog](../assets/supbase_setup_images/Supbase_6.png)

## Step 6.1: Copy Database Details to config.yaml

Copy the database connection details and add them to your `Week 4/config/config.yaml` file. Update the supabase section with:

- **host**: db.xxxxx.supabase.co (from the connection details)
- **port**: 5432
- **database**: postgres  
- **user**: postgres

![Copy to config.yaml](../assets/supbase_setup_images/Supbase_6_1.png)

## Verify Setup

Test your Supabase connection with the code below.

````python
# Quick test to verify your Supabase connection works
import os
import psycopg2
from dotenv import load_dotenv
from pathlib import Path
import yaml

# Load environment variables
load_dotenv("../.env")

# Load config
with open("../config/config.yaml", "r") as f:
    config = yaml.safe_load(f)

SUPABASE_PASSWORD = os.getenv("SUPABASE_PASSWORD")

if SUPABASE_PASSWORD:
    print("✅ SUPABASE_PASSWORD found in environment")
    
    # Build connection string
    supabase_config = config['supabase']
    DATABASE_URL = (
        f"postgresql://{supabase_config['user']}:{SUPABASE_PASSWORD}@"
        f"{supabase_config['host']}:{supabase_config['port']}/{supabase_config['database']}"
    )
    
    try:
        # Test connection
        conn = psycopg2.connect(DATABASE_URL)
        cur = conn.cursor()
        
        # Simple query to verify connection
        cur.execute("SELECT version();")
        version = cur.fetchone()[0]
        
        print("✅ Supabase connection successful!")
        print(f"PostgreSQL version: {version[:50]}...")
        
        cur.close()
        conn.close()
        
    except Exception as e:
        print(f"❌ Database connection failed: {e}")
else:
    print("❌ SUPABASE_PASSWORD not found. Please add it to your .env file.")
````
