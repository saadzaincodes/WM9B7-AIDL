# BGE-M3 Embedding Model Setup

This notebook downloads the `BAAI/bge-m3` model and verifies it can generate embeddings. The model will be cached locally after the first download.

## Step 1: Install Dependencies


## Step 2: Download & Verify the Model

Running the cell below will:
1. Download `BAAI/bge-m3` from HuggingFace on first run (~2 GB — this may take a few minutes)
2. Generate a test embedding to confirm everything is working

````python
from FlagEmbedding import BGEM3FlagModel

print("Loading BAAI/bge-m3 model (downloading if not cached)...")
model = BGEM3FlagModel("BAAI/bge-m3", use_fp16=True)
print("✅ Model loaded successfully!")

# Quick embedding test
test_sentence = "BGE-M3 supports dense, sparse, and ColBERT embeddings."
output = model.encode(
    [test_sentence],
    return_dense=True,
    return_sparse=True,
    return_colbert_vecs=True,
)

print(f"\n✅ Dense embedding shape:    {output['dense_vecs'].shape}")
print(f"✅ Sparse tokens (lexical):   {len(output['lexical_weights'][0])} non-zero entries")
print(f"✅ ColBERT vectors shape:     {output['colbert_vecs'][0].shape}")
print("\n🎉 BGE-M3 is ready to use!")
````
